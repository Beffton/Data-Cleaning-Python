```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Data_survey.xlsx")
pd.set_option('display.max_columns', None)

df.head()
```

```python
#Dropping columns that aren't needed for the project

df = df.drop(columns= ["Email", "Date Taken (America/New_York)", "Time Taken (America/New_York)", "Browser", "OS", "City", "Country", "Referrer", "Time Spent"])

df.head()
```

```python
#Simplifying column names

df.rename(columns= {"Q1 - Which Title Best Fits your Current Role?": "Job title",
                    "Q2 - Did you switch careers into Data?": "Swapped carrers into data",
                    "Q3 - Current Yearly Salary (in USD)": "Yearly salary range",
                    "Q4 - What Industry do you work in?": "Industry",
                    "Q5 - Favorite Programming Language": "Favorite progamming language",
                    "Q6 - How Happy are you in your Current Position with the following? (Salary)": "Satisfaction with current salary",
                    "Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)": "Satisfaction with current work/life balance",
                    "Q6 - How Happy are you in your Current Position with the following? (Coworkers)": "Satisfaction with current coworkers",
                    "Q6 - How Happy are you in your Current Position with the following? (Management)": "Satisfaction with current managment",
                    "Q6 - How Happy are you in your Current Position with the following? (Upward Mobility)": "Satisfaction with potential for upward mobility",
                    "Q6 - How Happy are you in your Current Position with the following? (Learning New Things)": "Satisfaction with potential for learning new things",
                    "Q7 - How difficult was it for you to break into Data?": "Difficulty for breaking into data",
                    "Q8 - If you were to look for a new job today, what would be the most important thing to you?": "Most important factor when looking for a new job",
                    "Q9 - Male/Female?": "Gender",
                    "Q10 - Current Age": "Age",
                    "Q11 - Which Country do you live in?": "Country of residence",
                    "Q12 - Highest Level of Education": "Highest level of education",
                    "Q13 - Ethnicity": "Ethnicity"}, 
                    inplace = True)

df.head()
```

```python
#taking care of NULL/empty values

df.isna().sum()
```

```python
#For columns with numbers, NULL values will be replaced by 0

df["Satisfaction with current salary"] = df["Satisfaction with current salary"].fillna(0)
df["Satisfaction with current work/life balance"] = df["Satisfaction with current work/life balance"].fillna(0)
df["Satisfaction with current coworkers"] = df["Satisfaction with current coworkers"].fillna(0)
df["Satisfaction with current managment"] = df["Satisfaction with current managment"].fillna(0)
df["Satisfaction with potential for upward mobility"] = df["Satisfaction with potential for upward mobility"].fillna(0)
df["Satisfaction with potential for learning new things"] = df["Satisfaction with potential for learning new things"].fillna(0)

#df.isna().sum()
```

```python
#For education, Null values will be replaced with "not discolsed"

#df["Highest level of education"].unique()

df["Highest level of education"] = df["Highest level of education"].fillna("Not disclosed")

df.isna().sum()
```

```python
#need to standarize columns with "other" as a response
#To keep things simple, I will be grouping all "other" responses into a single group

df["Job title"] = df["Job title"].str.replace(r"Other:.*", "Other", regex=True)
df["Industry"] = df["Industry"].str.replace(r"Other:.*", "Other", regex=True)
df["Favorite progamming language"] = df["Favorite progamming language"].str.replace(r"Other:.*", "Other", regex=True)
df["Most important factor when looking for a new job"] = df["Most important factor when looking for a new job"].str.replace(r"Other:.*", "Other", regex=True)
df["Country of residence"] = df["Country of residence"].str.replace(r"Other:.*", "Other", regex=True)

df
```

```python
#Making an age range column 

age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ["0-18", "19-25", "26-35", "36-45", "46-55", "56-65", "66+"]

df["Age range"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels, right=True)

df.head()
```

```python
#converting floats into integers for simpler visualization

df["Satisfaction with current salary"] = df["Satisfaction with current salary"].astype(int)
df["Satisfaction with current work/life balance"] = df["Satisfaction with current work/life balance"].astype(int)
df["Satisfaction with current coworkers"] = df["Satisfaction with current coworkers"].astype(int)
df["Satisfaction with current managment"] = df["Satisfaction with current managment"].astype(int)
df["Satisfaction with potential for upward mobility"] = df["Satisfaction with potential for upward mobility"].astype(int)
df["Satisfaction with potential for learning new things"] = df["Satisfaction with potential for learning new things"].astype(int)

df.head()
```

```python
#Making an "Average salary" column to make calculations easier

df["Salary temp"] = df["Yearly salary range"].str.replace("[k+]", "", regex=True)

df[["Minimum salary", "Maximum salary"]] = df["Salary temp"].str.split("-", expand=True)

df["Minimum salary"] = df["Minimum salary"].astype(int)
df["Maximum salary"] = df["Maximum salary"].fillna(df["Minimum salary"]).astype(int)

df["Average salary"] = (df["Minimum salary"] + df["Maximum salary"]) / 2

df.drop(columns= ["Salary temp", "Minimum salary", "Maximum salary"], inplace=True)

df.head()
```
